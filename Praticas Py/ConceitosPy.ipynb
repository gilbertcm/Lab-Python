```python
text = "Praticando"

print(text.lower()) # converte a string para min√∫sculo

print(text.upper()) # converte a string para mai√∫sculo

print("teste".upper()) # converte a string para mai√∫sculo

print(text.capitalize()) # converte primeiro caractere para mai√∫sculo

print(text.find("a")) # Mostra o √≠ndice do caractere

print(text.find("z")) # Retornar√° -1 se o caractere n√£o existir na string

print(" Ol√° Mundo ".strip()) #Remove os espa√ßos do inicio e fim, n√£o do meio

print("Python √© incrivel".split()) #Divide uma string em uma Lista de SubStrings
```

#### üìù Fun√ß√µes de Strings

*    *As fun√ß√µes upper() e lower() permitem que voc√™ mude rapidamente o caso de uma string para tudo em mai√∫sculas ou min√∫sculas, respectivamente.*

*    *Fun√ß√µes que funcionam apenas em certos objetos (strings, listas, etc.) s√£o chamadas usando a nota√ß√£o de ponto.<br> exemplo: "objeto.fun√ß√£o"*

*    *A fun√ß√£o capitalize() converte primeiro caractere de uma string para mai√∫sculas*

*    *A fun√ß√£o find() verifica se um caractere (ou um padr√£o de caracteres) est√° presente em uma string. A fun√ß√£o retorna o √≠ndice (posi√ß√£o) do valor fornecido. Se o valor fornecido estiver presente v√°rias vezes, a fun√ß√£o retornar√° a primeira ocorr√™ncia (o √≠ndice mais baixo).* <br>
*find() retornar√° -1 se o valor n√£o puder ser encontrado na string.*

```python
x = [16, 5, 3, 9, 7]

# Soma o valor n√∫merico
total = sum(x)
print("Soma: {}" .format(total))

# Soma a qtde de itens
total = len(x)
print(f"Quantidade de itens: {total}")

# Max() Retorna o maior valor da lista
print("Maior valor:", max(x))

# Retorna o menor valor
print("Menor valor:", min(x))

# Ordena os valores da lista
sorteio = sorted(x)
print(sorteio)

# Ordena de forma decrescente
sorteio = sorted(x, reverse=True)
print(sorteio)
```

#### üìù Fun√ß√µes incorporadas

* *A fun√ß√£o sum() recebe uma lista como argumento e soma todos os elementos da lista.*
* *A fun√ß√£o max() retorna o valor m√°ximo em uma lista.*
* *A fun√ß√£o min() retorna o valor m√≠nimo em uma lista.*
* *A fun√ß√£o sorted() recebe um iter√°vel como entrada e retorna uma list com os itens ordenados.*
* *A fun√ß√£o sorted() pode lidar tanto com valores num√©ricos quanto textuais. Para valores textuais, ela os ordena alfabeticamente.*
* *<b>Reverse</b> Quando reverse = True, os valores s√£o classificados em ordem decrescente.*

```python
# Cole√ß√£o imut√°vel
tupla = (16, "Gilbert", True, 1.78, 16, "Gilbert")

print("Ocorrencias do 16:", tupla.count(16))
print("Ocorrencias da String:", tupla.count("Gilbert"))
print("Ocorrencias do 1995:", tupla.count(1995))
print()

# Descompacta√ß√£o
data_anivers√°rio = (16, "mar√ßo", 1995)

dia, mes, ano = data_anivers√°rio
print(dia)
print(mes)
print(ano)
print()

dia, *resto = data_anivers√°rio
print(dia)
print(resto)
```

#### üìùTrabalhando com tuplas

* *A fun√ß√£o count() calcula o n√∫mero de ocorr√™ncias de um item em uma tupla.*
* *O operador * na descompacta√ß√£o de tuplas √© usado para reunir v√°rios elementos da tupla em uma lista. Isso √© √∫til ao lidar com tuplas de comprimento desconhecido.*

```python
# Conjuntos cole√ß√µes desordenadas mut√°veis

conjunto = {"Estudando", "Python", 2025}
conjunto2 = {"Uesc", 7 > 5, "Student", 2025}
# conjuntos = {"Duplicata", "Python", "Duplicata"} # Erro Duplicata 2x

print(conjunto)
# print(conjunto[0]) # Erro! n√£o suportam indexa√ß√£o ou fatiamento.

conjunto.add("Novo")
print(conjunto)

conjunto.remove("Novo")
print(conjunto)

# conjunto.append("Novo") # Erro! append s√≥ funciona em lista ordenadas

uniao = conjunto.union(conjunto2)
print(uniao)

diference = conjunto.difference(conjunto2)
print(diference)

conjunto.clear()
print(conjunto)
```

#### üìùConjuntos
*Cole√ß√µes desordenadas mut√°veis*


* *A fun√ß√£o clear() n√£o aceita um argumento e remove todos os itens de um conjunto.*
* *A fun√ß√£o union() retorna um novo conjunto com todos os elementos de ambos os conjuntos, omitindo duplicatas.*
* *A fun√ß√£o difference() retorna um conjunto contendo elementos que est√£o apenas no primeiro conjunto e n√£o no segundo.*




```python
# Dicion√°rio

product = {
    "name": "Caneta",
    "color": "red",
    "price": 2.99
}

# Indexa√ß√£o
print(product["name"])

produto = product.get("color")
print(produto)

info_keys = product.keys()
info_values = product.values()
print(info_keys)
print(info_values)

info = product.items()
print(info)

# Atribuindo valor para price
product["price"] = 1.99
print(product.items())

# Update atualiza o dicion√°rio
product.update({"name": "L√°pis"})
print(product.items())

product.update({"ano": "2025"})
print(product.items())

# Remove item usando a chave
product.pop("ano")
print(product.items())

# Verifica se chave ou valor esta no dicion√°rio
print("price" in product)
print()

# Imprime as chaves
for i in product:
    print(i)
    
print()
    
# Imprime os valores
for i in product.values():
    print(i)
```

#### üìùDicion√°rio

* *Dicion√°rios s√£o tipos de cole√ß√µes usados para armazenar dados em pares de chave:valor, que s√£o considerados como itens. Eles s√£o ideais para organizar dados em pares, onde cada peda√ßo de dado (valor) tem seu identificador √∫nico (chave).*
* *√≠nddice: Para acessar valores em dicion√°rios, precisa usar as keys.*
* *A fun√ß√£o get() acessa valores em um dicion√°rio*
* *A fun√ß√£o values() e keys() obtem todos os valores e chaves de um dicion√°rio*
* *A fun√ß√£o items() retorna todos os pares de chave:valor em um dicion√°rio.*
* *A fun√ß√£o update() mudar ou adicionar itens ao dicion√°rio.*



```python
# Compreens√£o de Listas
nums = [i for i in range(1, 21)]
print(nums)

# Dobra os valores
nums = [i * 2 for i in range(10)]
print(nums)
```
