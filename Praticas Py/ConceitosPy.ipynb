```python
lista = ["Brasil", "Itabuna", "Bahia"]

lista.append("New CIty") # Adiciona um item ao final de uma lista
print(lista)

lista.insert(0, "America") # Insere na lista na posição especifica
print(lista)

lista.pop(0) # Remove o item da lista na posição especifica
print(lista)

print("tamanho =", len(lista)) # Mostra o tam da lista,  qtde de itens na lista

print("tamanho =", len("Python")) # qtde de caractere
```

#### 📝 Anotações

*    *len() significa length e, quando usado em listas, retorna o número de itens na lista, e strings o número de caracteres*

*   *A função append() adiciona um novo item ao final de uma lista. é chamado usando a notação de ponto porque é específico para listas.*

*   *A função insert(posição, argumento) permite que você adicione um elemento a uma lista, em uma posição específica.*

*   *A função pop() remove um elemento de uma lista.*

```python
text = "Praticando"

print(text.lower()) # converte a string para minúsculo

print(text.upper()) # converte a string para maiúsculo

print("teste".upper()) # converte a string para maiúsculo

print(text.capitalize()) # converte primeiro caractere para maiúsculo

print(text.find("a")) # Mostra o índice do caractere

print(text.find("z")) # Retornará -1 se o caractere não existir na string
```

#### 📝 Funções de Strings

*    *As funções upper() e lower() permitem que você mude rapidamente o caso de uma string para tudo em maiúsculas ou minúsculas, respectivamente.*

*    *Funções que funcionam apenas em certos objetos (strings, listas, etc.) são chamadas usando a notação de ponto.<br> exemplo: "objeto.função"*

*    *A função capitalize() converte primeiro caractere de uma string para maiúsculas*

*    *A função find() verifica se um caractere (ou um padrão de caracteres) está presente em uma string. A função retorna o índice (posição) do valor fornecido. Se o valor fornecido estiver presente várias vezes, a função retornará a primeira ocorrência (o índice mais baixo).* <br>
*find() retornará -1 se o valor não puder ser encontrado na string.*
